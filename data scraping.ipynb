# selector exaple
- element element --> div p --> select all p elements inside div elements
- element > element --> div > p --> selects all p element where the parent is a div element
- element + element --> div + p --> selects first p element that is placed immediately after div elements
- element ∼ element --> p ∼ ul --> selects every ul element that are preceeded by p element
