# masking
- mask is used as a query in sql
- to select the distinct items from the dataset we use mask methodology

```python
countries_covid_data
```

```python
(countries_covid_data['Country'] == "India") 
```

```python
countries_covid_data['Date'] == '2021-06-22'
```

```python
mask = (countries_covid_data['Country'] == "India") & (countries_covid_data['Date'] == '2021-06-22')
```

```python
mask
```

```python
countries_covid_data[mask]
```

```python
mask = (countries_covid_data['Country'] == "Zimbabwe") & (countries_covid_data['Date'] == '2021-06-22')
```

```python
countries_covid_data[mask]
```

```python
# we can also do like this

countries_covid_data[(countries_covid_data['Country'] == "India") & (countries_covid_data['Date'] == '2021-06-22')]
```

```python
# new methodology is using query

countries_covid_data.query("(Country == 'Zimbabwe') | (Date == '2021-06-22')")
```

```python
# new methodology is using query

countries_covid_data.query("(Country == 'Zimbabwe') & (Date == '2021-06-22')")
```
