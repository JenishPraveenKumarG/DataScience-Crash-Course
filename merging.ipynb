# Merging dataframes
- 5 types <br>
- inner: only rows which are present on both dataframes are kept<br>
- left: row which are present on both dataframes are kept along with lefover rows from left dataframe<br>
- right: row which are present on both dataframes are kept along with lefover rows from right dataframe<br>
- outer join/ full join:  row which are present on both dataframes are kept along with lefover rows from right and left dataframe<br>
- anti-join: rows which are present on both dataframes are removed, keeping only the left over rows from both dataframe

```python
import pandas as pd
```

countries_covid = pd.read_csv('covid-19-main/countries-aggregated.csv')

```python
countries_covid.head()
```

```python
countries_ref = pd.read_csv('covid-19-main/reference.csv')
```

```python
countries_ref.head()
```

```python
countries_ref
```

```python
countries_ref_no_states = countries_ref[countries_ref['Province_State'].isna()]
```

```python
countries_ref_no_states
```

```python
countries_ref_no_states = countries_ref_no_states.rename(columns = {"Country_Region" : "Country"})
```

```python
countries_ref_no_states
```

```python
countries_covid.merge(countries_ref_no_states[['Population' , 'Country']])
```

```python
countries_covid_split
```
